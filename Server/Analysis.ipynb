```python
import pandas as pd
import numpy as np
import requests, json
import datetime
```

```python
df = pd.read_excel('DataBase.xlsx')
df
```

```python
df['timestamp'] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
df.columns = ['sensor_id', 'analog', 'mgl', 'solution', 'timestamp']
df
```

```python
df.solution.unique()
```

```python
data = [row.to_dict() for i, row in df.iterrows()]
data
```

```python
data[0]
```

```python
data = json.dumps(data, ensure_ascii=False)
data
```

```python
serverURL = r'https://sighir.com:8000/api/addSensors/'
response = requests.post(serverURL, data=data, timeout=100000)
response
```
