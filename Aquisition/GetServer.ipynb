```python
import pandas as pd
import requests
import json
```

```python
serverURL = 'https://sighir.com:8000/api/'
```

```python
def postData(url, route, data, timeout=15):
    url  = url + route
    data = json.dumps(data)
    
    try:
        response = requests.post(url, data, timeout=timeout)
        return response.json()
    except:
        return False


response = postData(serverURL, 'getRows/', {'tableName': 'sensors'})
response
```

```python
rows = response['data']
rows[:5]
```

```python
df = pd.DataFrame(rows)
df
```

```python
df.to_csv('DataBase.csv', index=None)
```
